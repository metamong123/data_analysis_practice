```python
def get_statistics(temp_list):
    """데이터의 기본 통계량 리턴"""
    min_value = min(temp_list)
    max_value = max(temp_list)
    avg_value = sum(temp_list)/len(temp_list)
    return min_value, max_value, avg_value
```

```python
get_statistics([185.5, 192.3, 179.2, 195.1])
```

```python
a, b, c = get_statistics([185.5, 192.3, 179.2, 195.1])
```

```python
print(a, b, c)
```

```python
def cal_sum(*numbers): # *: 파라미터를 튜플로 받기
    print(f'받은 파라미터: {numbers}')
    print(f'타입: {type(numbers)}')
    return sum(numbers)
```

```python
ret = cal_sum(2, 3, 5)
```

```python
# 아이디와 타입과 온도를 받아서 화면에 출력하는 함수
# 온도가 190 이상이면 'WARNING', 그렇지 않으면 'NORMAL' 표시도 함께
```

```python
def print_machine(**args):
    print(f"[STAT] {args['m_id']}({args['m_type']}): {'WARNING' if args['temp'] >= 190 else 'NORMAL'}")
```

```python
my_data = {'m_id': 'MC001', 'm_type': 'CNC', 'temp': 185.5}
```

```python
['MC001', 'CNC', 185.5]
```

```python
print_machine(m_id='MC001', m_type='CNC', temp=195.5)
```

```python
my_data = {'m_id': 'MC001', 'm_type': 'CNC', 'temp': 185.5, 'color': 'red'}
```

```python
def print_machine(m_id, m_type, temp, **kwargs):
    print(f"[STAT] {m_id}({m_type}): {'WARNING' if temp >= 190 else 'NORMAL'}")
```

```python
print_machine(**my_data)
```
