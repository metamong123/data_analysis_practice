```python
'''
1. 센서 온도 데이터 분석

temperatures = [185.5, 192.3, 178.2, 195.1, 182.4, 188.7, 201.2, 176.8]
다음 작업을 수행하시오:

최고 온도와 최저 온도 출력
평균 온도 계산
온도를 오름차순으로 정렬
190도 이상인 온도의 개수 세기

[예상 결과]
최고 온도: 201.2
최저 온도: 176.8
평균 온도: 188.15
정렬된 온도: [176.8, 178.2, 182.4, 185.5, 188.7, 192.3, 195.1, 201.2]
190도 이상: 2개
'''

temperatures = [185.5, 192.3, 178.2, 195.1, 182.4, 188.7, 201.2, 176.8]

print(f'최고 온도: {max(temperatures)}')
print(f'최저 온도: {min(temperatures)}')
print(f'평균 온도: {sum(temperatures)/len(temperatures)}')
print(f'정렬된 온도: {sorted(temperatures)}')
print(f'190도 이상: {len([temp for temp in temperatures if temp >= 190])}개')
```

```python
'''
2. 생산량 데이터 슬라이싱

c
다음 작업을 수행하시오:

첫 5개 데이터 추출
마지막 3개 데이터 추출
홀수 인덱스 데이터만 추출
역순으로 변환
처음과 마지막 2개씩 제외한 중간 데이터 추출

[예상 결과]
첫 5개: [120, 135, 128, 142, 138]
마지막 3개: [140, 136, 150]
홀수 인덱스: [135, 142, 145, 140, 150]
역순: [150, 136, 140, 132, 145, 138, 142, 128, 135, 120]
중간 데이터: [128, 142, 138, 145, 132, 140]
'''

production = [120, 135, 128, 142, 138, 145, 132, 140, 136, 150]

print(f'첫 5개: {production[:5]}')
print(f'마지막 3개: {production[-3:]}')
print(f'홀수 인덱스: {production[1::2]}')
print(f'역순: {production[::-1]}')
print(f'중간 데이터: {production[2:-2]}')
```

```python
'''
3. 교대조 불량 개수 처리

shift_A = [3, 5, 2, 4]
shift_B = [1, 2, 3, 2]
다음 작업을 수행하시오:

두 리스트 연결
shift_A를 3번 반복
연결된 리스트의 총 불량 개수
최대 불량 개수와 최소 불량 개수

[예상 결과]
연결: [3, 5, 2, 4, 1, 2, 3, 2]
반복: [3, 5, 2, 4, 3, 5, 2, 4, 3, 5, 2, 4]
총 불량: 22개
최대: 5개
최소: 1개
'''

shift_A = [3, 5, 2, 4]
shift_B = [1, 2, 3, 2]

print(f'연결: {shift_A + shift_B}')
print(f'반복: {shift_A*3}')
print(f'총 불량: {sum(shift_A + shift_B)}개')
print(f'최대: {max(shift_A + shift_B)}개')
print(f'최소: {min(shift_A + shift_B)}개')
```

```python
'''
4. 설비 ID와 온도 매핑

machine_ids = ["MC001", "MC002", "MC003", "MC004"]
temperatures = [185.5, 192.3, 178.2, 195.1]
다음 작업을 수행하시오:

두 리스트를 튜플 리스트로 결합
machine_ids에 인덱스 추가 (1부터 시작)
결합된 리스트에서 첫 번째와 마지막 튜플 출력

[예상 결과]
결합: [('MC001', 185.5), ('MC002', 192.3), ('MC003', 178.2), ('MC004', 195.1)]
인덱스 추가: [(1, 'MC001'), (2, 'MC002'), (3, 'MC003'), (4, 'MC004')]
첫 번째: ('MC001', 185.5)
마지막: ('MC004', 195.1)
'''

machine_ids = ["MC001", "MC002", "MC003", "MC004"]
temperatures = [185.5, 192.3, 178.2, 195.1]

print(f"결합: {[(machine_ids[idx], temperatures[idx]) for idx in range(len(machine_ids))]}")
print(f"결합: {[(m_id, temp) for m_id, temp in zip(machine_ids, temperatures)]}")
print(f"인덱스 추가: {[(idx + 1, machine_ids[idx]) for idx in range(len(machine_ids))]}")
print(f"인덱스 추가: {[(idx + 1, m_id) for idx, m_id in enumerate(machine_ids)]}")
print(f"첫 번째: {[(machine_ids[idx], temperatures[idx]) for idx in range(len(machine_ids))][0]}")
print(f"마지막: {[(machine_ids[idx], temperatures[idx]) for idx in range(len(machine_ids))][-1]}")
```

```python
'''
5. 중첩 리스트 센서 데이터

sensor_data = [
    [185.5, 186.2, 184.8, 185.0, 186.1],
    [120.3, 121.5, 119.8, 120.1, 120.9],
    [45.2, 44.8, 45.5, 45.1, 44.9]
]
다음 작업을 수행하시오:

첫 번째 센서의 평균 계산
두 번째 센서의 최댓값과 최솟값
세 번째 센서의 첫 3개 측정값
모든 센서의 첫 번째 측정값을 리스트로 생성

[예상 결과]
센서 1 평균: 185.52
센서 2 최댓값: 121.5
센서 2 최솟값: 119.8
센서 3 첫 3개: [45.2, 44.8, 45.5]
모든 센서 첫 번째 값: [185.5, 120.3, 45.2]
'''

sensor_data = [
    [185.5, 186.2, 184.8, 185.0, 186.1],
    [120.3, 121.5, 119.8, 120.1, 120.9],
    [45.2, 44.8, 45.5, 45.1, 44.9]
]

print(f"센서 1 평균: {sum(sensor_data[0])/len(sensor_data[0])}")
print(f"센서 2 최댓값: {max(sensor_data[1])}")
print(f"센서 2 최솟값: {min(sensor_data[1])}")
print(f"센서 3 첫 3개: {sensor_data[2][:3]}")
print(f"모든 센서 첫 번째 값: {[sensor[0] for sensor in sensor_data]}")
```

```python
'''
6. 에러 코드 리스트 조작

errors = [101, 102, 103]
다음 작업을 수행하시오:

104 추가
[105, 106] 추가
인덱스 2에 201 삽입
마지막 요소 제거 및 출력
101의 개수 세기
최종 리스트의 길이 출력

[예상 결과]
104 추가 후: [101, 102, 103, 104]
[105, 106] 추가 후: [101, 102, 103, 104, 105, 106]
201 삽입 후: [101, 102, 201, 103, 104, 105, 106]
제거된 요소: 106
101 개수: 1개
최종 길이: 6
'''

errors = [101, 102, 103]

errors.append(104)
print(f"104 추가 후: {errors}")
errors.extend([105, 106])
print(f"[105, 106] 추가 후: {errors}")
errors.insert(2, 201)
print(f"201 삽입 후: {errors}")
print(f"제거된 요소: {errors.pop()}")
print(f"101 개수: {errors.count(101)}개")
print(f"최종 길이: {len(errors)}")
```

```python
'''
8. 확장 언패킹

measurements = (185.5, 186.2, 184.8, 185.0, 186.1, 187.3, 188.2)
다음 작업을 수행하시오:

첫 번째 값과 나머지를 분리
첫 번째, 마지막, 중간 값들을 분리
나머지와 마지막 값을 분리
각 변수의 타입과 값 출력

[예상 결과]
first: 185.5, rest: [186.2, 184.8, 185.0, 186.1, 187.3, 188.2]
first: 185.5, middle: [186.2, 184.8, 185.0, 186.1, 187.3], last: 188.2
beginning: [185.5, 186.2, 184.8, 185.0, 186.1, 187.3], last: 188.2
rest 타입: <class 'list'>
middle 타입: <class 'list'>
'''

measurements = (185.5, 186.2, 184.8, 185.0, 186.1, 187.3, 188.2)
first, *rest = measurements
print(f"first: {first}, rest: {rest}")
first, *middle, last = measurements
print(f"first: {first}, middle: {middle}, last: {last}")
*beginning, last = measurements
print(f"beginning: {beginning}, last: {last}")
print(f"rest 타입: {type(rest)}")
print(f"middle 타입: {type(middle)}")
```

```python
'''
9. 튜플 연산

coords_A = (10, 20, 30)
coords_B = (40, 50)
다음 작업을 수행하시오:

두 튜플 연결
coords_A를 3번 반복
30이 coords_A에 있는지 확인
coords_A의 길이와 최댓값, 최솟값

[예상 결과]
연결: (10, 20, 30, 40, 50)
반복: (10, 20, 30, 10, 20, 30, 10, 20, 30)
30 in coords_A: True
길이: 3
최댓값: 30
최솟값: 10
'''

coords_A = (10, 20, 30)
coords_B = (40, 50)

print(f"연결: {coords_A + coords_B}")
print(f"반복: {coords_A*3}")
print(f"30 in coords_A: {30 in coords_A}")
print(f"길이: {len(coords_A)}")
print(f"최댓값: {max(coords_A)}")
print(f"최솟값: {min(coords_A)}")
```

```python
'''
10. 중첩 딕셔너리 설비 정보

machine = {
    "id": "MC001",
    "type": "CNC",
    "location": "Building A",
    "sensors": {
        "temperature": 185.5,
        "pressure": 1.2,
        "vibration": 0.05
    },
    "status": "Running"
}
다음 작업을 수행하시오:

전체 딕셔너리 출력
온도 값만 출력
센서 데이터에 'current' 항목 추가 (값: 15.5)
상태를 'Maintenance'로 변경
최종 딕셔너리 출력

[예상 결과]
온도: 185.5
최종 상태: Maintenance
최종 센서 데이터: {'temperature': 185.5, 'pressure': 1.2, 'vibration': 0.05, 'current': 15.5}
'''

machine = {
    "id": "MC001",
    "type": "CNC",
    "location": "Building A",
    "sensors": {
        "temperature": 185.5,
        "pressure": 1.2,
        "vibration": 0.05
    },
    "status": "Running"
}

print(machine)
print(f"초기 센서 데이터: {machine['sensors']}")
print(f"온도: {machine['sensors']['temperature']}")
machine['sensors']['current'] = 15.5
machine['status'] = "Maintenance"
print(f"최종 상태: {machine['status']}")
print(f"최종 센서 데이터: {machine['sensors']}")
print(machine)
```
