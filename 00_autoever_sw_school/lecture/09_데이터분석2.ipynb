```python
df_production['불량률'] = round(df_production['defect_quantity']/df_production['actual_quantity']*100, 2)
```

```python
# 불량률이 5프로 미만이면 '우수'
# 그렇지 않으면 '주의'라고 표시하는 등급 칼럼 만들기
```

```python
import numpy as np
```

```python
df_production['등급'] = np.where(df_production['불량률'] < 5, '우수', '주의')
```

```python
df_production
```

```python
# np.select()
# 3가지 이상 조건으로 분류할 때
```

```python
# 불량률이 3보다 작으면 S등급
# 7보다 작으면 A등급
# 10보다 작으면 B등급
# 그렇지 않으면 C등급
```

```python
conditions = [df_production['불량률'] < 3, df_production['불량률'] < 7, df_production['불량률'] < 10]
choices = ['S등급', 'A등급', 'B등급']
default = 'C등급'
```

```python
df_production['등급'] = np.select(conditions, choices, default)
```

```python
df_production
```

```python
### pd.apply() 방식
df_production['불량률']
```

```python
def get_grade_bin(defact_rate):
    return '우수' if defact_rate < 5 else '주의'
```

```python
df_production['등급'] = df_production['불량률'].apply(get_grade_bin)
```

```python
df_production
```

```python
def get_grade_multi(defact_rate):
    if defact_rate < 3: return 'S등급'
    elif defact_rate < 7: return 'A등급'
    elif defact_rate < 10: return 'B등급'
    else: return 'C등급'
```

```python
df_production['등급'] = df_production['불량률'].apply(get_grade_multi)
```
