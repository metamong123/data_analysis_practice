## 문제 6: 다중 조건 필터링 (20점)

`production_df`에서:
1. `equipment_id`가 'INJ-001'이고 `defect_quantity`가 5개 이상인 데이터를 선택하세요
2. `actual_quantity`가 50개 미만이거나 150개 초과인 데이터를 선택하세요
3. `product_code`가 'BUMPER-A'이고 `good_quantity`가 100개 이상인 데이터의 개수를 출력하세요



```python
# 1. INJ-001 설비 AND 불량 5개 이상
production_df[(production_df['equipment_id'] == 'INJ-001') & (production_df['defect_quantity'] >= 5)]
```

```python
# 2. 생산량 50개 미만 OR 150개 초과
production_df[(production_df['actual_quantity'] < 50) | (production_df['actual_quantity'] > 150)]
production_df[~(production_df['actual_quantity'].between(50, 150))]
```

```python
# 3. BUMPER-A 제품 AND 양품 100개 이상
len(production_df[(production_df['product_code'] == 'BUMPER-A') & (production_df['good_quantity'] >= 100)])
```

1. `equipment_id`가 'INJ-001', 'INJ-002', 'INJ-003' 중 하나인 데이터를 선택하고 건수를 출력하세요
2. `actual_quantity`가 80~120개 사이인 데이터를 선택하고 처음 5개를 출력하세요



```python
# 1. INJ-001, INJ-002, INJ-003 설비
production_df[production_df['equipment_id'].isin(['INJ-001', 'INJ-002', 'INJ-003'])]
```

## 도전 문제: 품질 검사 데이터 분석 (보너스 20점)

`../data/07_quality_inspection.csv` 파일을 읽어서 다음을 분석하세요:

1. 전체 검사 건수를 출력하세요
2. `result`가 'FAIL'인 불량 데이터만 선택하세요
3. 불량률(불량 건수 / 전체 건수 * 100)을 계산하여 출력하세요
4. `equipment_id`가 'INJ-001'이고 `result`가 'FAIL'인 데이터의 개수를 출력하세요
5. `measurement_value`(측정값)가 300 미만이거나 310 초과인 데이터를 선택하세요 (규격 이탈)

**힌트**: 위에서 배운 모든 내용을 종합적으로 활용하세요!

```python
# 1. 전체 검사 건수를 출력하세요
q_df = pd.read_csv("../data/07_quality_inspection.csv")
q_df
```

```python
# 2. `result`가 'FAIL'인 불량 데이터만 선택하세요
q_df[q_df['result'] == 'FAIL']
```

```python
# 3. 불량률(불량 건수 / 전체 건수 * 100)을 계산하여 출력하세요
len(q_df[q_df['result'] == 'FAIL'])/len(q_df)*100
```

```python
# 4. `equipment_id`가 'INJ-001'이고 `result`가 'FAIL'인 데이터의 개수를 출력하세요
len(q_df[(q_df['equipment_id'] == 'INJ-001') & (q_df['result'] == 'FAIL')])
```

```python
# 5. `measurement_value`(측정값)가 300 미만이거나 310 초과인 데이터를 선택하세요 (규격 이탈)
q_df[(q_df['measurement_value'] < 300) | (q_df['measurement_value'] > 310)]
```
