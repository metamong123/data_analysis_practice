```python
'''
1. 변수 선언 및 할당

# 다음 정보를 적절한 변수명으로 저장하시오:

설비 이름: "CNC Milling Machine"
설비 ID: "MC001"
현재 온도: 185.5
목표 온도: 180.0
가동 상태: True
에러 코드: None

[예상 결과]
설비 이름: CNC Milling Machine
설비 ID: MC001
현재 온도: 185.5
목표 온도: 180.0
가동 상태: True
에러 코드: None
'''

op_name = "CNC Milling Machine"
op_id = "MC001"
op_curtemp = 185.5
op_targettemp = 180.0
op_activate = True
op_errcode = None

print(f'설비 이름: {op_name}')
print(f'설비 ID: {op_id}')
print(f'현재 온도: {op_curtemp}')
print(f'목표 온도: {op_targettemp}')
print(f'가동 상태: {op_activate}')
print(f'에러 코드: {op_errcode}')
```

```python
'''
2. 여러 변수 동시 할당

다음 값들을 한 줄에 변수에 할당하시오:

temperature, pressure, humidity에 각각 185.5, 1.2, 45 할당
두 변수 a, b의 값을 교환 (a=10, b=20 → a=20, b=10)

[예상 결과]
temperature: 185.5
pressure: 1.2
humidity: 45
교환 전 - a: 10, b: 20
교환 후 - a: 20, b: 10

'''

temperature, pressure, humidity = 185.5, 1.2, 45

print(f'temperature: {temperature}')
print(f'pressure: {pressure}')
print(f'humidity: {humidity}')

a, b = 10, 20
print(f'교환 전 - a: {a}, b: {b}')
a, b = b, a
print(f'교환 후 - a: {a}, b: {b}')
```

```python
'''
3. 박스 포장 계산

total_items = 1250
items_per_box = 48
총 생산량을 박스당 개수로 나눈 박스 수와 남는 개수를 계산하시오.

[예상 결과]
필요한 박스: 26개
남는 개수: 2개
'''

total_items = 1250
items_per_box = 48

print(f'필요한 박스: {total_items//items_per_box}개')
print(f'남는 개수: {total_items%items_per_box}개')
```

```python
'''
4. 생산 효율 계산

actual_production = 850
target_production = 1000
work_hours = 8
다음을 계산하시오:

목표 대비 생산율 (백분율)
시간당 평균 생산량
목표 달성에 필요한 추가 생산량

[예상 결과]
생산율: 85.0%
시간당 생산량: 106.25개
부족 수량: 150개
'''

actual_production = 850
target_production = 1000
work_hours = 8

print(f'생산율: {actual_production/target_production*100}%')
print(f'시간당 생산량: {actual_production/work_hours}개')
print(f'부족 수량: {target_production - actual_production}개')
```

```python
'''
5. 온도 단위 변환

celsius = 185.5
섭씨 온도를 화씨로 변환하시오. (공식: F = C × 9/5 + 32)

[예상 결과]
섭씨: 185.5°C
화씨: 365.9°F
'''

celsius = 185.5

print(f'섭씨: {celsius}°C')
print(f'화씨: {celsius*9/5 + 32}°F')
```

```python
'''
6. 복합 대입 연산자

count = 100
다음 작업을 순서대로 수행하고 각 단계의 결과를 출력하시오:

50 증가
2배 증가
30 감소
3으로 나눈 몫

[예상 결과]
초기값: 100
50 증가: 150
2배 증가: 300
30 감소: 270
3으로 나눈 몫: 90
'''

count = 100
print(f'초기값: {count}')
count += 50
print(f'50 증가: {count}')
count *= 2
print(f'2배 증가: {count}')
count -= 30
print(f'30 감소: {count}')
count //= 3
print(f'3으로 나눈 몫: {count}')
```

```python
'''
7. 거듭제곱과 나머지 연산

base = 2
exponent = 10
number = 1024
divisor = 7
다음을 계산하시오:

2의 10승
1024를 7로 나눈 몫과 나머지

[예상 결과]
2^10 = 1024
1024 ÷ 7 = 몫 146, 나머지 2
'''
base = 2
exponent = 10
number = 1024
divisor = 7

print(f'{base}^{exponent} = {base**10}')
print(f'{number} ÷ {divisor} = 몫 {number//divisor}, 나머지 {number%divisor}')
```

```python
'''
8. 문자열 인덱싱

machine_id = "MC-2024-A001"
다음 값을 추출하시오:

첫 번째 문자
마지막 문자
연도 부분 (2024)
뒤에서 4자리 (A001)

[예상 결과]
첫 문자: M
마지막 문자: 1
연도: 2024
뒤 4자리: A001
'''

machine_id = "MC-2024-A001"

print(f'첫 문자: {machine_id[0]}')
print(f'마지막 문자: {machine_id[-1]}')
print(f'연도: {machine_id.split("-")[1]}')
print(f'뒤 4자리: {machine_id[-4:]}')
```

```python
'''
9. 문자열 슬라이싱

serial = "FACTORY-MACHINE-12345"
다음을 추출하시오:

앞 7글자
뒤 5글자
중간 부분 (MACHINE)
역순으로 변환

[예상 결과]
앞 7글자: FACTORY
뒤 5글자: 12345
중간 부분: MACHINE
역순: 54321-ENIHCAM-YROTCAF
'''

serial = "FACTORY-MACHINE-12345"

print(f'앞 7글자: {serial[:7]}')
print(f'뒤 5글자: {serial[-5:]}')
print(f'중간 부분: {serial.split("-")[1]}')
print(f'역순: {serial[::-1]}')
```

```python
'''
10. 문자열 연결과 반복

prefix = "MC"
number = "001"
separator = "-"
다음을 생성하시오:

설비 ID: prefix + separator + number
구분선 40개의 등호(=)
"OK"를 5번 반복하고 쉼표로 구분

[예상 결과]
설비 ID: MC-001
========================================
OK,OK,OK,OK,OK
'''

prefix = "MC"
number = "001"
separator = "-"

print(f'설비 ID: {prefix + separator + number}')
print('='*40)
print(','.join(['OK']*5))
```

```python
'''
11. 문자열 포맷팅 비교

item = "Bolt"
quantity = 1500
price = 25.5
% 연산자, format() 메서드, f-string 세 가지 방법으로 동일한 결과를 출력하시오.

[예상 결과]
제품: Bolt, 수량: 1,500개, 가격: 25.50원
제품: Bolt, 수량: 1,500개, 가격: 25.50원
제품: Bolt, 수량: 1,500개, 가격: 25.50원
문자열 메서드
'''

item = "Bolt"
quantity = 1500
price = 25.5

print('제품: ' + item + ', 수량: ' + f'{quantity:,}' + '개, 가격: ' + f'{price:.2f}' + '원')
print('제품: {}, 수량: {:,}개, 가격: {:.2f}원'.format(item, quantity, price))
print(f'제품: {item}, 수량: {quantity:,}개, 가격: {price:.2f}원')
```

```python
'''
12. 대소문자 변환

product_code = "SmArT-FaCtOrY-2024"
다음으로 변환하시오:

모두 대문자
모두 소문자
첫 글자만 대문자
각 단어의 첫 글자만 대문자

[예상 결과]
대문자: SMART-FACTORY-2024
소문자: smart-factory-2024
첫 글자 대문자: Smart-factory-2024
제목 형식: Smart-Factory-2024
'''

product_code = "SmArT-FaCtOrY-2024"

print(f'대문자: {product_code.upper()}')
print(f'소문자: {product_code.lower()}')
print(f'첫 글자 대문자: {product_code.capitalize()}')
print(f'제목 형식: {product_code.title()}')
```

```python
'''
13. 공백 제거 및 정렬

log_message = "   System Error 404   "
item_name = "Bolt"
다음을 수행하시오:

양쪽 공백 제거
왼쪽만 제거
오른쪽만 제거
item_name을 20자리 왼쪽 정렬
item_name을 20자리 오른쪽 정렬

[예상 결과]
양쪽 제거: 'System Error 404'
왼쪽 제거: 'System Error 404   '
오른쪽 제거: '   System Error 404'
왼쪽 정렬: 'Bolt                '
오른쪽 정렬: '                Bolt'
'''

log_message = "   System Error 404   "
item_name = "Bolt"

print(f"양쪽 제거: '{log_message.strip()}'")
print(f"왼쪽 제거: '{log_message.lstrip()}'")
print(f"오른쪽 제거: '{log_message.rstrip()}'")
print(f"왼쪽 정렬: '{item_name.ljust(20)}'")
print(f"왼쪽 정렬: '{item_name:<20}'")
print(f"오른쪽 정렬: '{item_name.rjust(20)}'")
print(f"오른쪽 정렬: '{item_name:>20}'")
```

```python
'''
14. 문자열 검색

log = "Machine MC001 temperature high at 14:30:25"
다음을 수행하시오:

"temperature" 문자열의 위치 찾기
"MC001" 문자열의 위치 찾기
문자 'e'의 등장 횟수 세기
"Machine"으로 시작하는지 확인
"25"로 끝나는지 확인

[예상 결과]
temperature 위치: 14
MC001 위치: 8
e 등장 횟수: 5
Machine으로 시작: True
25로 끝남: True
'''

log = "Machine MC001 temperature high at 14:30:25"

print(f'temperature 위치: {log.find("temperature")}')
print(f'MC001 위치: {log.find("MC001")}')
print(f'e 등장 횟수: {log.count("e")}')
print(f'Machine으로 시작: {log.startswith("Machine")}')
print(f'25로 끝남: {log.endswith("25")}')
```

```python
'''
15. 문자열 치환 및 분리

log = "ERROR: Temperature exceeded at sensor-1"
data = "MC001,185.5,Running,Normal"
다음을 수행하시오:

"ERROR"를 "WARNING"으로 변경
"sensor-1"을 "sensor-2"로 변경
data를 쉼표로 분리하여 리스트로 만들기
분리된 리스트를 " | "로 결합

[예상 결과]
변경 후: WARNING: Temperature exceeded at sensor-2
분리: ['MC001', '185.5', 'Running', 'Normal']
결합: MC001 | 185.5 | Running | Normal
'''

log = "ERROR: Temperature exceeded at sensor-1"
data = "MC001,185.5,Running,Normal"

print(f"변경 후: {log.replace('ERROR', 'WARNING').replace('sensor-1', 'sensor-2')}")
print(f"분리: {data.split(',')}")
print(f"결합: {' | '.join(data.split(','))}")
```

```python
'''
16. 문자열 판별

code1 = "12345"
code2 = "MC001"
code3 = "ABCDE"
code4 = "   "
다음을 확인하시오:

code1이 숫자로만 구성되었는지
code2가 문자와 숫자로 구성되었는지
code3가 문자로만 구성되었는지
code4가 공백으로만 구성되었는지

[예상 결과]
code1 숫자: True
code2 문자+숫자: True
code3 문자: True
code4 공백: True
'''

code1 = "12345"
code2 = "MC001"
code3 = "ABCDE"
code4 = "   "

print(f'code1 숫자: {code1.isdigit()}')
print(f'code2 문자+숫자: {code2.isalnum()}')
print(f'code3 문자: {code3.isalpha()}')
print(f'code4 공백: {code4.isspace()}')
```

```python
'''
17. f-string 고급 포매팅

machine_name = "CNC Machine"
current_temp = 185.5
target_temp = 180.0
production = 1500
efficiency = 0.856
다음 형식으로 출력하시오:

온도: 소수점 1자리
생산량: 천 단위 구분
효율: 백분율 (소수점 1자리)
온도 편차: 부호 포함

[예상 결과]
[설비 상태 보고서]
설비명: CNC Machine
현재 온도: 185.5°C
목표 온도: 180.0°C
온도 편차: +5.5°C
생산량: 1,500개
효율: 85.6%
'''

machine_name = "CNC Machine"
current_temp = 185.5
target_temp = 180.0
production = 1500
efficiency = 0.856

print(f'''[설비 상태 보고서]
설비명: {machine_name}
현재 온도: {current_temp}°C
목표 온도: {target_temp}°C
온도 편차: {"+" if current_temp - target_temp > 0 else "-"}{abs(current_temp - target_temp)}°C
생산량: {production:,}개
효율: {efficiency*100}%''')
```
