```python
# 어떤 설비가 가장 많이 사용되는가?
# 생산 부하가 집중된 설비 파악 -> 병목 지점
df_production.groupby('equipment_id').size().sort_values(ascending=False)
df_production['equipment_id'].value_counts()
```

```python
# 어떤 설비가 가장 많은 제품을 생산했는가?
# 고성능 설비를 알 수 있음 -> 생산 계획시 우선 배정
df_production.groupby('equipment_id')['actual_quantity'].sum().sort_values(ascending=False)
```

```python
# 한 번 작업할 때 평균적으로 얼마나 생산하는가?
df_production.groupby('equipment_id')['actual_quantity'].mean().round(2).sort_values(ascending=False)
```

```python
# 각 설비별 평균 생산량, 표준편차(안정성), 최소, 최댓값, 개수
df_summary = df_production.groupby('equipment_id')['actual_quantity'].agg(['mean', 'std', 'min', 'max', 'count']).round(2)
```

```python
df_summary.rename(columns={'mean': '평균', 'std':'표준편차', 'min':'최솟값', 'max':'최댓값', 'count':'개수'})
```

```python
df_production.columns
```

```python
# 생산성 지표, 양품 생산 능력, 불량 발생 정도, 품질 수준
# 생산성은 높은데, 불량률도 높은 것을 찾을 수 있음 -> 품질 개선 프로젝트 진행

# df_production.groupby('equipment_id').agg(생산건수=('actual_quantity', 'count'))
df_production.groupby('equipment_id').agg({'actual_quantity':['sum', 'mean'],
                                           'good_quantity': 'sum',
                                           'defect_quantity': 'sum',
                                           'defect_rate': 'mean'})
```

```python
# 설비 + 제품별로 생산량 확인 -> 설비별 제품특화 파악
# 교대조 + 설비별로 불량률 확인 -> 주간조와 야간조의 차이 발견
```

```python
# 각 설비는 어떤 제품을 주로 생산하는가? -> 설비별로 제품 특화 파악, 교체작업 빈도 예측, 생산스케쥴링을 최적화 가능
df_production.groupby(['equipment_id', 'product_code']).agg(생산건수=('production_id', 'count'),
                                                            총생산량=('actual_quantity', 'sum'),
                                                            평균불량률=('defect_rate', 'mean')).round({'평균불량률': 2})
```

```python
# 설비별 평균 생산량을 각 행에 추가
# df_production.groupby('equipment_id')['actual_quantity'].mean()
df_production['설비별_평균_생산량'] =  df_production.groupby('equipment_id')['actual_quantity'].transform('mean').round(2)
```

```python
df_production.head()
```

```python
# 평균 대비 편차 -> 실제 생산량 - 평균 생산량
df_production['평균_대비_편차'] = (df_production['actual_quantity'] - df_production['설비별_평균_생산량']).round(2)
```

```python
df_production
```

```python
# Z-score: (데이터 수치 - 평균 수치)/표준편차
# 이상치 탐지할 때 사용(IQR, Z-Score)
# Z-score > 3 : 이상치
# Z-score > 2.56 : 신뢰구간 99%
# Z-score > 1.96 : 신뢰구간 95%
```

```python
df_production['설비별_표준편차_생산량'] = df_production.groupby('equipment_id')['actual_quantity'].transform('std')
```

```python
df_production['qty_zscore'] = df_production['평균_대비_편차']/df_production['설비별_표준편차_생산량']
```

```python
df_production
```

```python
# z-score가 3보다 큰 데이터 선택
df_production[df_production['qty_zscore'] > 3] # 이상치 없음
```

```python
# 생산 건수가 100건 미만인 설비는 빼고 선택
def data_size(x):
    return True if len(x) >= 100 else False

df_production.groupby('equipment_id').filter(data_size)
```

```python
# 평균 불량률이 5% 이상인 데이터만 선택
def data_mean(x):
    return True if x['defect_rate'].mean() >= 5 else False

df_production.groupby('equipment_id').filter(data_mean)
```
