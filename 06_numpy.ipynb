```python
# 이상치 탐지(IQR 방법)
```

```python
data = np.array([185, 186, 184, 250, 183, 187, 50, 185, 186, 184])
```

```python
q1 = np.percentile(data, 25)
```

```python
q3 = np.percentile(data, 75)
```

```python
iqr = q3 - q1
```

```python
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
```

```python
data[(data < lower_bound) | (data > upper_bound)]
```
