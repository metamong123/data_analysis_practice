```python
# 설비가 INJ-001이면서 불량이 5개 이상인 데이터
df_production[(df_production['equipment_id'] == 'INJ-001') & (df_production['defect_quantity'] >= 5)]
```

```python
# 평균 생산량
# 평균 양품 갯수
# 평균 불량 겟수
# 평균 불량률(%): [불량 개수]/[전체 개수]*100
```

```python
df_target = df_production[(df_production['equipment_id'] == 'INJ-001') & 
                          (df_production['product_code'] == 'BUMPER-A')]
```

```python
df_target.shape
```

```python
df_target.head()
```

```python
print(f"평균 생산량: {df_target['actual_quantity'].mean():.1f}개")
```

```python
df_target['good_quantity'].mean()
```

```python
df_target['defect_quantity'].mean()
```

```python
print(f"평균 불량률: {df_target['defect_quantity'].sum()/df_target['actual_quantity'].sum()*100:.2f} %")
```
